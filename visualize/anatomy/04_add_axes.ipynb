```python
# auto-scale
fig = plt.figure(figsize=(10,10), facecolor='linen')

left, bottom = 0.1, 0.1
spacing = 0.05
width1, height1 = 0.5, 0.5

width2 = 1 - (2 * left + width1 + spacing)
height2 = 1 - (2 * bottom + height1 + spacing)

rect1 = [left, bottom, width1, height1]
rect2 = [left + width1 + spacing, bottom, width2, height1]
rect3 = [left, bottom + height1 + spacing, 1 - 2 * left, height2]

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```

```python
# auto-scale 2
def auto_scale(left, bottom, spacing, width, height):
    width2 = 1 - (2 * left + width1 + spacing)
    height2 = 1 - (2 * bottom + height1 + spacing)

    rect1 = [left, bottom, width1, height1]
    rect2 = [left + width1 + spacing, bottom, width2, height1]
    rect3 = [left, bottom + height1 + spacing, 1 - 2 * left, height2]
    
    return rect1, rect2, rect3


fig = plt.figure(figsize=(10,10), facecolor='linen')

left, bottom = 0.1, 0.1
spacing = 0.005
width1, height1 = 0.65, 0.65

rect1, rect2, rect3 = auto_scale(left, bottom, spacing, width1, height1)

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```
