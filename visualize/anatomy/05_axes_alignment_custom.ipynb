```python
import matplotlib.pyplot as plt
```

### 1. fig.tight_layout : Axes Layout Adjustment
    - set_title
    - set_xlabel
    - set_ylabel

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_title('Title!', fontsize=20)
ax.set_xlabel('X label!', fontsize=15)
ax.set_ylabel('Y label!', fontsize=15)
```

```python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['X label ' + str(i) for i in range(4)]
ylabel_list = ['Y label ' + str(i) for i in range(4)]

print(title_list)
print(xlabel_list)
print(ylabel_list)

fig, axes = plt.subplots(2, 2, figsize=(10,10))
for idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[idx], fontsize=30)
    ax.set_xlabel(xlabel_list[idx], fontsize=20)
    ax.set_ylabel(ylabel_list[idx], fontsize=20)
```

#### fig.tight_layout(pad argument)

```python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['X label ' + str(i) for i in range(4)]
ylabel_list = ['Y label ' + str(i) for i in range(4)]

print(title_list)
print(xlabel_list)
print(ylabel_list)

fig, axes = plt.subplots(2, 2, figsize=(10,10))
for idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[idx], fontsize=30)
    ax.set_xlabel(xlabel_list[idx], fontsize=20)
    ax.set_ylabel(ylabel_list[idx], fontsize=20)
    
fig.tight_layout(pad=3)
```

### 2. fig.subplots_adjust : More Customized Layout
    - ax.get_x/yaxis() : 축 제거

```python
fig, ax = plt.subplots(figsize=(10,10))

x_axis = ax.get_xaxis()
y_axis = ax.get_yaxis()

x_axis.set_visible(False)
y_axis.set_visible(False)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))

for idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(left=0.3, bottom=0.3, right=0.8, top=0.7)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))

for idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(hspace=0.05, wspace=0.05)
```
