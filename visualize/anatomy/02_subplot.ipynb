```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10), facecolor='gray')
axes[0, 0].plot([1, 2, 3])
axes[0, 1].plot([3, 2, 1])
axes[1, 0].plot([4, 5, 6])
axes[1, 1].plot([6, 5, 4])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10), facecolor='linen')
title_list = ['first', 'second', 'third', 'fourth']
for i, ax in enumerate(axes.flat):
    ax.set_title(title_list[i], fontsize=15)
    ax.plot([i, (i+1), (i+2)])
```
