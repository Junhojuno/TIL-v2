### Subplot2grid
    - plt.subplot : 정형화된 패턴으로 밖에 사용하지 못함
    - add_suplot : plt.subplot보다는 불규칙한 패턴에 유연하지만, 한계가 있음

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,10), facecolor='linen')

ax1 = plt.subplot2grid(shape=(2,1), loc=(0,0), fig=fig)
ax1.set_title('ax1', fontsize=15)

ax2 = plt.subplot2grid(shape=(2,1), loc=(1,0), fig=fig)
ax2.set_title('ax2', fontsize=15)
```

```python
fig = plt.figure(figsize=(10,10), facecolor='linen')

ax1 = plt.subplot2grid(shape=(1,2), loc=(0,0), fig=fig)
ax1.set_title('ax1', fontsize=15)

ax2 = plt.subplot2grid(shape=(1,2), loc=(0,1), fig=fig)
ax2.set_title('ax2', fontsize=15)
```

```python
fig = plt.figure(figsize=(10,10), facecolor='linen')

ax1 = plt.subplot2grid(shape=(3,3), loc=(0,0), colspan=2, title='ax1', fig=fig)
ax2 = plt.subplot2grid(shape=(3,3), loc=(1,0), colspan=3, title='ax2', fig=fig)
ax3 = plt.subplot2grid(shape=(3,3), loc=(2,0), title='ax3', fig=fig)
ax4 = plt.subplot2grid(shape=(3,3), loc=(0,2), title='ax4', fig=fig)
ax5 = plt.subplot2grid(shape=(3,3), loc=(2,1), colspan=2, title='ax5', fig=fig)
```

```python
fig = plt.figure(figsize=(10,10))

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3, title='ax1', fig=fig)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2, title='ax2', fig=fig)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2, title='ax3', fig=fig)
ax4 = plt.subplot2grid((3,3), (2,0), title='ax4', fig=fig)
ax5 = plt.subplot2grid((3,3), (2,1), title='ax5', fig=fig)
```

```python
fig = plt.figure(figsize=(20,20), facecolor='skyblue')

ax1 = plt.subplot2grid((5,4), (0,0), rowspan=2, colspan=2, title='ax1', fig=fig)
ax2 = plt.subplot2grid((5,4), (0,2), rowspan=2, colspan=2, title='ax2', fig=fig)

count = 3
for i in range(2, 5):
    for j in range(4):
        title = f'ax{count}'
        ax = plt.subplot2grid((5,4), (i, j), title=title, fig=fig)
        count += 1
```
